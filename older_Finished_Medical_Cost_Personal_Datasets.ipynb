```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import colorsys
plt.style.use('seaborn-talk')
import warnings
warnings.filterwarnings('ignore')

sns.set_style()
%matplotlib inline
```

```python
# Importando o arquivo como df
df = pd.read_csv('insurance.csv')

# Lendo as primeiras entradas
df.head()
```

**Dicionário das colunas:**

- `age`: idade de cada consumidor do seguro;

- `sex`: sexo do consumidor;

- `bmi`: *body mass index*. No Brasil, é o índice de massa corpórea, **IMC**;

- `children`: se o consumidor tem filhos ou não;

- `smoker`: se o consumidor fuma ou não;

- `region`: região que o consumidor mora;

- `charges`: custos que o consumidor teve/tem com o seguro de saúde.

```python
# Shape do dataset
print(f"Variáveis: {df.shape[1]}\nDados inseridos: {df.shape[0]}")
```

```python
# Verificando quantos dados nulos existem
df.isnull().sum()
```

A célula anterior nos mostrou que não existem valores nulos no nosso conjunto de dados. Para ficar mais fácil, vamos renomear as colunas

```python
df.rename(columns={'age': 'idade', 'sex': 'sexo', 'bmi': 'imc', 'children': 'filhos', 'smoker': 'fumante', 'region': 'regiao', 'charges': 'custos'}, inplace=True)
```

```python
# Visualizando o tipo dos dados
print(df.dtypes, '\n')

# Visualizando DataFrame
df
```

É muito comum em problemas amostrais dentro da estatística, a presença de valores **discrepantes**, ou melhor, **_outliers_**. Estes valores fogem totalmente do "padrão" do dataset e muitas vezes pode ser um problema, uma vez que interfere nos resultados dos teses.

Vamos verificar a presença de *outliers*.

```python
df.describe()
```

Fazendo uma leitura abaixo do resultado da célula acima, temos o seguinte:

> **Obs.:** lembre-se que existem variáveis binárias (sexo e fumante), portanto alguns valores não devem existir.

- **count**: contagem dos valores, ou seja, **quantos valores existem. Como todos possuem a mesma quantidade, sabemos que não existem dados faltantes em nenhuma das colunas.

- **mean**: se refere à média dos valores. Podemos utilizá-la para fins de comparação quando existir a hipótese de *outlier*.

- **std**: representa o **desvio padrão**, o quão dispersos os dados estão da média.

- **min**: menor valor.

- **25%**: 1/4 do quartil.

- **50%**: 1/2 do quartil, **mediana**.

- **75%**: 3/4 do quartil.

- **max**: 4/4 do quartil, maior valor.

*Existem outliers?*

A priori, os dados não parecem estar discrepantes. Não ao ponto de induzir a nossa análise ao erro.

Os números da variávei `filhos` está bem menor comparado aos números das outras variáveis, o que mais para frente pode ser levado em consideração.

As colunas `idade` e `imc` estão, relativamente, próximas umas as outras, exceto pela `mean`, `max`, `75%` e `std`.

Vamos visualizar as distribuições das variáveis `idade`, `imc` e `custos`.


```python
fig, axs = plt.subplots(ncols = 3, figsize = (20,6));
sns.distplot(df['idade'], ax = axs[0]);
sns.distplot(df['imc'], ax = axs[1]);
sns.distplot(df['custos'], ax = axs[2]);
```

De cara, percebemos que a variável `imc` é dada por uma distribuição normal. A variável `custos`, aparentemente é uma highly skewed. Já na primeira variável, `idade`, podemos ver que quem predomina são as pessoas na faixa dos 20 anos, seguida por 40 e uns 63, 65, respectivamente.

Dado que temos variáveis categóricas, vamos plotar gráficos a fim de visualizar como é a distribuição dentre elas.

```python
fig2, axs2 = plt.subplots(ncols = 2, figsize = (20,8));
sns.countplot(df['sexo'], ax = axs2[0]);
sns.countplot(df['regiao'], ax = axs2[1]);
```

Podemos visualizar algumas informações como:

- a quantidade de homens é **ligeiramente** maior que mulheres;

- a população de *southeast* gasta mais que as outras.

Porém, como não são grandes diferenças, podemos ver que os dados estão bem distribuídos ao longo das categorias.

Vamos visualizar como as variáveis categóricas da célula anterior afetam a variável `custo`.

```python
fig3, axs3 = plt.subplots(ncols = 1, figsize = (14, 8))
sns.boxplot(x = df['regiao'], y = df['custos'], hue = df['sexo'], ax = axs3);
```

Pelo gráfico acima, é possível que existam outliers e que variáveis como `sexo` e `regiao` não trazem impactos aos `custos`. 

É comum se pensar que ter filhos necessariamente aumenta as despesas médicas. Portanto, vamos visualizar o impacto da quantidade de filhos aos custos médicos.

```python
fig4, axs4 = plt.subplots(ncols = 1, figsize = (16, 8))
sns.boxplot(x = df['filhos'], y = df['custos'], ax = axs4);
```

Similar ao gráfico anterior, podemos ter outliers, porém não é claro se existe um impacto significativo.

Vejamos o impacto de fumantes nos custos.

```python
fig5, axs5 = plt.subplots(ncols = 1, figsize = (12, 8))
sns.boxplot(x = df['fumante'], y = df['custos'], ax = axs5);
```

No gráfico acima, vemos que existe um impacto significativo entre fumante e os custos.

Vejamos como se comporta a relação entre fumantes, IMC e idade em relação aos custos.

```python
fig6, axs6 = plt.subplots(ncols = 2, figsize = (15,10))
sns.scatterplot(x = df[df['fumante'] == 'yes']['idade'], y = df[df['fumante'] == 'yes']['custos'], ax = axs6[0]).set_title("Fumantes");
sns.scatterplot(x = df[df['fumante'] == 'no']['idade'], y = df[df['fumante'] == 'no']['custos'], ax = axs6[1]).set_title("Não-Fumantes");
```

Os gráficos acima têm algumas características:

**Gráfico Fumantes:**

- relação positiva entre a idade e os custos, porém existe um *gap*.

**Gráfico Não-Fumantes:**

- relação positiva nentre a idade e os custos;

- uma parte do gráfico tem uma dispersão **muito baixa**;

- outra parte do gráfico tem alta dispersão.

Conclui-se que existe outra variável que está influenciando os dados, mas ainda não conseguimos visualizar. Dado que este problema é de natureza de saúde, podemos relacionar a variável **imc**, **filhos**, além da já inclusa (**fumantes**).

Vimos que a variável **imc** é uma curva de sino (vide distribuição normal). Assim, vamos separá-la em um `IMC > 30` e um `IMC < 30`.

```python
# Separando a variavel imc
df['sobrepeso'] = np.where(df['imc'] > 30, 'yes', 'no')

# Visualizando a nova coluna
df['sobrepeso']
```

```python
# Relacionando a nova variavel sobrepeso
fig7, axs7 = plt.subplots(ncols = 2, figsize = (16, 8))
sns.scatterplot(x = df[df['fumante'] == 'yes']['idade'], y = df[df['fumante'] == 'yes']['custos'], ax = axs7[0], hue = df['sobrepeso']).set_title('Fumante');
sns.scatterplot(x = df[df['fumante'] == 'no']['idade'], y = df[df['fumante'] == 'no']['custos'], ax = axs7[1], hue = df['sobrepeso']).set_title('Não-Fumante');
```

Especialmente o gráfico de fumantes, vemos que a influência da variável `sobrepeso` (feita para quem tem imc > 30) é **muito significativa** a ponto de termos uma conclusão.

> **Pessoas que fumam e têm sobrepeso arcam com mais despesas médicas ao longo da vida.**

Já no segundo gráfico, não está claro se, de fato, existe uma influência pelo sobrepeso. Desta forma, devemos avaliar outra variável. Os passos abaixo são repetições do passo anterior apenas mudando a variável `sobrepeso` no gráfico de não-fumantes.


```python
# Avaliando influência das outras variáveis
fig8, axs8 = plt.subplots(ncols = 3, figsize = (16,8))
sns.scatterplot(df[df['fumante'] == 'no']['idade'], y = df[df['fumante'] == 'no']['custos'], hue = df['filhos'], ax = axs8[0]).set_title("Não-Fumantes e Filhos");
sns.scatterplot(df[df['fumante'] == 'no']['idade'], y = df[df['fumante'] == 'no']['custos'], hue = df['regiao'], ax = axs8[1]).set_title("Não-Fumantes e Região");
sns.scatterplot(df[df['fumante'] == 'no']['idade'], y = df[df['fumante'] == 'no']['custos'], hue = df['sexo'], ax = axs8[2]).set_title("Não-Fumantes e Sexo");
```

Ainda assim, não nos parece claro como um grupo de não-fumantes tem maiores custos que outro grupo de não-fumantes. Como o nosso estoque de variáveis é pequeno, com elas não conseguimos afirmar como isso acontece.
